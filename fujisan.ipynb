```python
"""
【課題】富士山を超える最小の折る回数
"""
PAPER = 0.00008
MT_FUJI_HIGHT = 3776
paper_hight = PAPER
fold_count = 0

while True:
    if paper_hight > MT_FUJI_HIGHT:
        break
    else:
        fold_count += 1
        paper_hight *= 2
        print("{}回折った時に、{}メートルになり富士山の高さを超える！！".format(fold_count,paper_hight))

print("{}回折った時に、{:.1f}メートルになり富士山の高さを超える！！".format(fold_count,paper_hight))
# print(paper_hight)
```

```python
"""
【課題】任意の厚さに対応した関数を作成
t0 は PAPER
参考URL：
http://aoki2.si.gunma-u.ac.jp/Hanasi/StatTalk/fudoushousuu.html
e+16の意味→10の１６乗

"""
def manual_thickness(input_thickness,one_thickness):

#     初期処理（一枚の厚さ、折った紙の高さ、折った回数）
    PAPER = one_thickness
    paper_hight = PAPER
    fold_count = 0

#     任意の高さを超えるまで繰り返し、折った回数のカウント
    while True:
        if input_thickness < paper_hight:
            break
        else:
            fold_count += 1
            paper_hight *= 2
            print ("{}折ると、高さは{}になる。".format(fold_count,paper_hight))

    pass
    return fold_count,paper_hight

# 任意の高さを入力する。
print("任意の高さをメートルで入力してね！！^ - ^")
print("最も近い太陽以外の恒星はプロキシマ・ケンタウリといい、距離は4.243光年で、4.01419e+16メートル！！^ - ^；")
print("つまり、、401419000000000000000メートル！！^ - ^；")
input_hight = input()

# 紙の厚さの初期値を入力する。
print("紙の厚さを入力してね！！！（通常 0.00008m）")
input_thickness = input()

# 入力した値は文字型になっているので数値型にして引数を渡す
# returnされた返り値を左辺に取得
fold_count , paper_hight = manual_thickness(float(input_hight),float(input_thickness))

# 返り値を出力
print ("{}折ると、高さは{}になる。".format(fold_count,paper_hight))
```

```python
"""
【課題】必要な紙の長さ
L = (pai * t0) / 6 *((2**n) + 4)*((2**n) - 1)

参考URL：
https://python.atelierkobato.com/pi/

月：43回折る→n=43
富士山：26回折る→n=26
最も近い太陽以外の恒星：83回折る→n=83
"""

def how_length_paper(fold_time):
    # 初期処理
    # pythonで用意されているπを用意
    import math
    pai = math.pi
    #     print("pai = {}".format(pai))
    PAPER = 0.00008

    # 折る回数は判明しているので、これを元に必要な長さを算出する
    paper_length = (pai * PAPER) / 6 *((2**fold_time) + 4)*((2**fold_time) - 1)

    pass
    return paper_length

# 折った回数を入力し、引数として関数に渡す
# 月：43回折る→n=43
# 富士山：26回折る→n=26
# 最も近い太陽以外の恒星：83回折る→n=83
print("紙を折る回数を入力してください〜m(_ _)m")
fold_time = input()
paper_length = how_length_paper(float(fold_time))

print("紙の長さは{:.6f}mです！".format(paper_length))
```
