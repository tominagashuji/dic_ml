```python
"""
【問題】栗まんじゅうが太陽系を覆う日
饅頭の体積：直径5cm（0.05m）の球
太陽の体積：直径140万kmの球
球体の体積の求め方：V=4/3*π*(r**3)
"""
import math
pai = math.pi

MANJU_RADIUS = 0.025
SUN_RADIUS = 75*(10**7)
                 
def wharf_covers_the_sun():

    volume = 0
    minutes = 0
    manju_volume = (4/3)*pai*(MANJU_RADIUS**3)
    sun_volume = (4/3)*pai*(SUN_RADIUS**3)
    print(manju_volume)
    print(sun_volume)
    while True:
        if manju_volume  > sun_volume:
            break
        else:
            minutes += 5
            manju_volume *= 2
            print("時間{}、饅頭体積{}".format(minutes,manju_volume))

    pass
    return minutes

minutes = wharf_covers_the_sun()
print("饅頭が太陽を覆うのは{}分後です！！！( ；∀；)<ｺﾜｲｰ".format(minutes))
```
