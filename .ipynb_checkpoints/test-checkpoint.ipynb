# 線形回帰のメモ・動作確認

```python
import numpy as np
import pandas as pd
import random as rd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
theta = np.array([rd.random() for i in range(4)])
print(theta.shape)
X = np.array([[-1, 2, 3, 4], [4, -5, 6, 7], [7, 8, -9, 10]])
X
```

```python
def _linear_hypothesis(self, X):
    """
    線形の仮定関数を計算する

    Parameters
    ----------
    X : 次の形のndarray, shape (n_samples, n_features)
      学習データ

    Returns
    -------
      次の形のndarray, shape (n_samples, 1)
      線形の仮定関数による推定結果

    """
    #(サンプル数 x 特徴量数) @ (特徴量数 x 1) = (サンプル数 x 1)
    h = np.dot(X, self)
    pass
    return h
```

```python
h = _linear_hypothesis(theta, X)
h
```

```python
h.shape
```

```python
y = np.array([rd.random() for i in range(3)])
print("y：{}".format(y))
y
```

```python
def _get_error(self, X):
    """
    説明を記述
    """
    error = h - y
    return error
```

```python
error = _get_error(theta, y)
error
```

```python
def _gradient_descent(theta, X, error):
    """
    説明を記述
    """
    m = X.shape[0]
#     print(error.shape)
#     print(X.T.shape)
    theta = theta - (0.01 / m *(X.T@error).sum())
    return theta
```

```python
theta = _gradient_descent(theta, X, error)
print(theta)
```

```python
def predict(theta, X):
    """
    線形回帰を使い推定する。

    Parameters
    ----------
    X : 次の形のndarray, shape (n_samples, n_features)
        サンプル

    Returns
    -------
        次の形のndarray, shape (n_samples, 1)
        線形回帰による推定結果
    """
    print(X.shape)
    print(theta.shape)
    #(サンプル数 x 特徴量数) @ (特徴量数 x 1) = (サンプル数 x 1)
    y_pred = np.dot(X, theta)

    pass
    return y_pred
```

```python
y_pred = predict(theta, X)
print(y_pred)
```

```python
def MSE(y_pred, y):
    """
    平均二乗誤差の計算

    Parameters
    ----------
    y_pred : 次の形のndarray, shape (n_samples,)
      推定した値
    y : 次の形のndarray, shape (n_samples,)
      正解値

    Returns
    ----------
    mse : numpy.float
      平均二乗誤差
    """
    m = y.shape[0]
    mse = ((y_pred - y)**2).sum()/m
    return mse
```

```python
mse = MSE(y_pred, y)
print(type(mse))
print(mse)
```
