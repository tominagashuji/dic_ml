```python
"""
問１：紙を1回折った時の厚さを計算するコード
"""

THICKNESS = 0.00008

folded_thickness = THICKNESS*2**43 # 43も変数にしても良い

print("厚さ： {}メートル".format(folded_thickness))
```

```python
"""
問2：メートルをキロメートルに変換して小数点以下2桁で表示する
"""
# メートルをキロメートルに変換して小数点以下2桁で表示する
print("厚さ： {:.2f}キロメートル".format(folded_thickness/1000))
```

```python
"""
問 3：for文で作成
"""
### for文で作成
THICKNESS = 0.00008
folded_thickness = THICKNESS

for i in range(43):# やんわり他の書き方がある
  folded_thickness *=2　# やってないけど THICKNESS （定数）を変更するのはやめよう
#   print(folded_thickness) 
    
print("厚さ： {:.2f}キロメートル".format(folded_thickness/1000))
```

```python
"""
問4：時間計測
"""
### for文の方が時間がかかる。

import time
start = time.time()
#####
THICKNESS = 0.00008
folded_thickness = THICKNESS

for i in range(43):
  folded_thickness *=2
#   print(folded_thickness) 
    
print("厚さ： {:.2f}キロメートル".format(folded_thickness/1000))
#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))



import time
start = time.time()
#####
THICKNESS = 0.00008

folded_thickness = THICKNESS*2**43

print("厚さ： {}メートル".format(folded_thickness)) #printは時間がかかる（計算よりもよっぽど時間掛かる）
#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))

"""
%%timeit というのがjupyterに存在して簡単にたくさん実行してくれる
べき乗の方が早い。forもprintも遅いよ。
"""
```

```python
"""
問5：リストへの保存
"""
### for文で作成
THICKNESS = 0.00008
folded_thickness = THICKNESS
folded_thickness_list = [0]

for i in range(43):
  folded_thickness *=2
  folded_thickness_list.append(folded_thickness)
#   print(folded_thickness) 
    
print(folded_thickness_list)
len(folded_thickness_list)
```

```python
"""
【問題6】折れ線グラフの表示
"""
### for文で作成
THICKNESS = 0.00008
folded_thickness = THICKNESS
folded_thickness_list = [0]

for i in range(43):
  folded_thickness *=2
  folded_thickness_list.append(folded_thickness)

"""
グラフを表示する。タイトルと軸ラベル名付き。
"""
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(folded_thickness_list) # 「リスト名」のところに問５のリストの変数名を入れる
plt.show()

##約40回目を折った辺りから急激に厚くなる
```

```python
"""
【問題7】グラフのカスタマイズ
"""
plt.plot(folded_thickness_list, color='pink')
plt.show()

plt.plot(folded_thickness_list, color='blue')
plt.show()

plt.plot(folded_thickness_list, color='green', marker='o', linestyle='dashed',
         linewidth=2, markersize=12)
plt.show()


```
