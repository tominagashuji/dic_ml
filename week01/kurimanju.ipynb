```python
"""
【問題】栗まんじゅうが太陽系を覆う日
饅頭の体積：直径5cm（0.05m）の球
太陽系の半径：4,498,252,900km
球体の体積の求め方：V=4/3*π*(r**3)
"""
# 計算で必要なπを使用できるようにする
import math
pai = math.pi

# 饅頭と太陽の半径を算出（メートルで統一）
MANJU_RADIUS = 0.025
SOLAR_SYSTEM_RADIUS = 4498252900 * 1000

def wharf_covers_the_sun():

    #     初期処理を実施
    minutes = 0
    
    #     饅頭と太陽の体積を算出
    manju_volume = (4/3)*pai*(MANJU_RADIUS**3)
    solar_system_volume = (4/3)*pai*(SOLAR_SYSTEM_RADIUS**3)
    #     念のため体積を確認
    #     print(manju_volume)
    #     print(solar_system_volume)

    #     饅頭の体積が太陽を超えるまで繰り返し実施
    while True:
        if manju_volume  > solar_system_volume:
            break
        else:
            #     饅頭は5分毎に倍に大きくなる
            minutes += 5
            manju_volume *= 2
            print("時間{}、饅頭体積{}".format(minutes,manju_volume))

    return minutes,manju_volume

minutes,manju_volume = wharf_covers_the_sun()
print("饅頭が太陽を覆うのは{}分後です！！！( ；∀；)<ｺﾜｲｰ！！！大きさは{:.1f}立方メートルです！！！".format(minutes,manju_volume))
```
