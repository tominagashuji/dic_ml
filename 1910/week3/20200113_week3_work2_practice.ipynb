````python
#列毎の欠損値の個数確認
null_count = df.isnull().sum() # 欠損値を数える
null_rate = (null_count/len(df)*100) #  欠損値を１００分率に変更する

# データフレームを作成
null_tbl = pd.DataFrame({
    'Total': null_count,
    'null_rate': null_rate
    })

# 欠損値があるものだけ抜き出し欠損率でソート
null_tbl[null_tbl['Total'] > 0].sort_values(by='null_rate', ascending=False)
````

#### 5つ以上欠損値がある特徴量（列）は削除する。
[参考URL](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html "https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html")

[参考URL_dropnaとか](https://qiita.com/0NE_shoT_/items/8db6d909e8b48adcb203 "https://qiita.com/0NE_shoT_/items/8db6d909e8b48adcb203")

````python
# 5つ以上欠損値がある特徴量（列）は削除する。

# 欠損値が5つ以上あるインデックスを抽出
null_tbl = null_tbl[null_tbl.Total >= 5].sort_values(by='null_rate', ascending=False)
delete_column = null_tbl.index # テーブルの行名(削除したい列名を取得

deleted_df1 = df.drop(delete_column, axis = 1) # 取得した列名をデータフレームから削除
deleted_df1.head()
````

#### 欠損値があるサンプル（行）は削除する。

````python
deleted_df2 = deleted_df1.dropna(how='any')
deleted_df2.head()
````

### 【問題5】用語の調査
以下の２つの単語を知る必要があります。これらを調査してまとめてください。

#### 尖度
>尖度（せんど、英: kurtosis)は、確率変数の確率密度関数や頻度分布の鋭さを表す数字である。  
正規分布と比べて、尖度が大きければ鋭いピークと長く太い裾を持った分布を持ち、  
尖度が小さければより丸みがかったピークと短く細い尾を持った分布であるという事が判断できる。

#### 歪度
>「歪度」（わいど）とは、分布の歪み度合いを示すものです。   
正規分布のように左右対称であるのか、それとも片方の裾だけだ伸びたような歪んだ分布なのか、その度合いを示します。   
これを歪度（または歪度係数）といって、歪度が0 よりも大きければ右の裾が長い分布、  
歪度が0 よりも小さければ左の裾が長い分布となります。


### 【問題6】分布の確認
seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。  
目的変数に対して対数変換を行う。

対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。

````python
sns.distplot(deleted_df2.SalePrice.values) # numpyで投入する必要あり。それとカラム指定をする必要もある。
plt.show()
````

````python
#歪度(Skewness)と尖度(Kurtosis)
print("歪度(Skewness) :", deleted_df2.SalePrice.skew())
print("尖度(Kurtosis) :", deleted_df2.SalePrice.kurt())
````

#### 目的変数に対して対数変換を行う。
[対数変換の方法](https://qiita.com/ynakayama/items/2c3782c82317f21596ba "https://qiita.com/ynakayama/items/2c3782c82317f21596ba")

````python
# data.apply(np.log)
deleted_df2.SalePrice.apply(np.log).head()
````

#### 対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

````python
sns.distplot(deleted_df2.SalePrice.apply(np.log)) # numpyで投入する必要あり。それとカラム指定をする必要もある。
plt.show()
````

```
機械学習をするときは正規分布に従ったデータの方が取り扱いやすい。
また、対数変換すると正規分布に近くことがほとんどなので、対数変換が一般的。
```

````python
print("歪度(Skewness) :", deleted_df2.SalePrice.apply(np.log).skew())
print("尖度(Kurtosis) :", deleted_df2.SalePrice.apply(np.log).kurt())
````

#### 対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。

>対数変換を実施前は尖度が高く（尖っている）、歪度が左によっている分布だった。  
それに対し、対数変換後は尖度が低く（尖りが緩く）、歪度が中央により、正規分布に近づいた。

### 【問題7】相関係数の確認
特徴量の確認をしていきます。それぞれに対してマークダウンで説明や考察を書いてください。

#### 相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。

````python
df_corr = deleted_df2.corr(method='pearson')
# print(deleted_df2_corr)

colormap = plt.cm.RdBu# 色の指定
plt.figure(figsize=(100, 15))# サイズの指定
sns.heatmap(df_corr, linewidths=0.1,vmax=1.0, 
             fmt='.1f',
             xticklabels=df_corr.columns.values,
             yticklabels=df_corr.columns.values,
             square=True, cmap=colormap, linecolor='white', annot=True)
plt.show()
````

#### 【考察】
```
GrageCarsとGarageArea、1stFlrSFとTotalBsmtSF、TotRmsAbvGrdとGrLivArea、は相関係数が高い。
EnclosedPorch、35snPorch、ScreenPorch、PoolArea、MiscVal、MoSold、YrSoldは他の項目と相関関係はほとんどない。
```

#### 目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。

````python
# 相関係数の高い順に行と列に目的変数でソートをかける。axis＝０は省略されている
df_corr_10 = df_corr.sort_values(by='SalePrice', ascending=False).sort_values(by='SalePrice', axis=1, ascending=False)
df_corr_10 = df_corr_10.iloc[:11, :11]#目的変数を含めたトップ１０を抽出
df_corr_10

# 相関係数行列のヒートマップ
plt.figure(figsize=(10, 5))
sns.heatmap(df_corr_10,
           vmin=-1.0,
           vmax=1.0,
            center=0,
            annot=True,#格子の中に値を表示させる
            fmt='.1f',
            xticklabels=df_corr_10.columns.values,
            yticklabels=df_corr_10.columns.values
           )
plt.show()
````
