### ANDゲートの学習データを用意

```python
import numpy as np
x_train = np.array([[0,0],[0,1],[1,0],[1,1]])
y_train = np.array([[0],[0],[0],[1]])
```

### Sequentialモデル

```python
import tensorflow as tf
# 全結合のクラス、tf.keras.layers.Denseを使う
model = tf.keras.Sequential([tf.keras.layers.Dense(1, activation = tf.nn.sigmoid, input_shape=(2,))])
```

### 作成したモデルの構造はsummaryメソッドで確認が出来る

```python

model.summary()
```

### 構造が記述できたらモデルをコンパイルする

```python
model.compile(loss='binary_crossentropy',# ここでは2値分類のため、binary_crossentropyとなります。
              optimizer=tf.train.AdamOptimizer(learning_rate=0.01),
              metrics=['accuracy'])
# 多値分類の場合はcategorical_crossentropy、回帰の場合はmean_squared_error
```

### 学習

```python
history = model.fit(x_train, y_train,
                   batch_size=1,
                   epochs=1000,
                   verbose=1)
```

検証用データがある場合は、引数validation_dataを与える事でエポック毎の検証が可能。

```python
history = model.fit(x_train, y_train,
                   batch_size=1,
                   epochs=1000,
                   verbose=1,
                   validation_data=(x_train, y_train))
```

### 推定はpredict

```python
y_pred_proba = model.predict(x_train)[:, 0]
# 確率を0, 1に変換
y_pred = np.where(y_pred_proba > 0.5, 1,0)
print("y_pred_proba", y_pred_proba)
print("y_pred", y_pred)
```

### 評価のみはevaluate

```python
score = model.evaluate(x_train, y_train, verbose=0)
print('Train loss', score[0])
print('Train accuracy: ', score[1])
```
