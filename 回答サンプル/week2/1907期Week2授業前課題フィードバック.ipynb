どういう場合に動くのか、どうだとダメなのかを意識しておくことは大切です！

```
#関数は(n,m)が引数 :n*mマスの盤
# returnは、n*mマスの盤の１＊{２＊＊(n＊m)}のarray

def chessboard_1(n,m): #n*mのチェス盤
    grains = 1
    count = 1
    board_list  = [1]
    
    number_of_boxes = n * m
   
    if number_of_boxes < 1: #マスの数が１より小さい時False
        return False
    
    elif type(number_of_boxes) == float: #nとmが整数じゃない時False
        return False
    
    else:
        for i in range(number_of_boxes - 1):
            board_list.append (2 * board_list[-1])
            
        #print("n*mマスの板に小麦を並べる（リスト）:{}".format(board_list))
        
        board_ndarray = np.array(board_list)
        
        #print("n*mマスの板に小麦を並べる（ndarray）:{}".format(board_ndarray))
        
        board_n_m_array = board_ndarray.reshape(n, m)
            
    return board_n_m_array, board_n_m_array.shape #配列とその形を返す


print(chessboard_1 (8, 8))
```

Pythonのfor文では`row`や`col`を明示的に`row += 1`のようにする必要はありません。


```
def func_naiseki_manual3(def_array1, def_array2):
        def_result_array = np.zeros(def_array1.shape[0]*def_array2.shape[1]).reshape(def_array1.shape[0],def_array2.shape[1])
        if def_array1.shape[1] != def_array2.shape[0]:
            return print("Can't calcurate. Because array1's columns don't equal to array2's raws") 
        else:
            row=0
            col=0
            for row in range(def_array1.shape[0]):
                for col in range(def_array2.shape[1]):
                    for kou in range(def_array1.shape[1]):
                        def_result_array[row,col] += def_array1[row,kou]*def_array2[kou,col]
                    col += 1
                row +=1
            return def_result_array

d_ndarray_ = np.array([[-1, 2, 3], 
                       [4, -5, 6]])
e_ndarray = np.array([[-9, 8, 7], 
                      [6, -5, 4]])
result_array = func_naiseki_manual3(d_ndarray_, e_ndarray)
result_array
```
