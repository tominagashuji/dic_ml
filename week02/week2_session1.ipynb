### 【問題1】1次関数
x が-50から50の範囲で次の式を考えます。
y=1/2x+1
-50から50まで0.1間隔のxを表すndarrayと、各要素に対するyを表すndarrayを作成してください。

```python
import numpy as np
```

```python
# 開始位置、終了位置、差分でarange可能
x = np.arange(-50, 50.1, 0.1)
# print(x)
y = 1/2*x + 1
# print(y)
print(x.shape)
print(y.shape)
```

### 【問題2】配列の結合
問題1で作成したndarrayを結合し、(1001, 2)の形の配列にしてください。0列目にx、1列目にyが来るようにします。

```python
# x を 1001行１列に変更する
x_r = x.reshape(1001,1)
```

```python
# y を 1001行１列に変更する
y_r = y.reshape(1001,1)
```

```python
# axis 1 方向で結合する
x_y_r = np.concatenate([x_r,y_r],1)
print(x_y_r)
```

```python
# オブジェクトのアトリビュートに代入するの方法
x.shape = (1001, 1)
x.shape
```

### 【問題3】勾配を求める
各x座標における勾配を求めます。ここで勾配はyの変化量をxの変化量で割ることで計算します。

```python
koubai = np.empty(1)
koubai[0] = 0.5
for i in range(len(x_y_r) - 1):
    # xの変化量を出力
    x_k = x_y_r[i+1, 0] - x_y_r[i, 0]
#     print(x_k)
    # yの変化量を出力
    y_k = x_y_r[i+1, 1] - x_y_r[i, 1]
#     print(y_k)
    # xの変化量をyの変化量で除算すると勾配が求まる
    koubai = np.append(koubai,[y_k / x_k]) # numpy を追加したら numpy に再度入れる必要があった！！
#     print(koubai)
print(len(x_y_r))
print(len(koubai))
print(koubai)
print(koubai.shape)
print(type(koubai))
```

```python
# 他のやり方
# 変化量d
d = x_y_r[1:] - x_y_r[:-1]
# 勾配slope
s = d[:, 1] / d[:,0]
print(s)
```

### 【問題4】グラフを描く
1次関数のグラフをMatplotlibを用いて描いてください。また、それとは別に横軸をx、縦軸を勾配としたグラフも描いてください。

```python
# 一次関数グラフ
import matplotlib.pyplot as plt
plt.title("y = 1/2*x + 1")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x_y_r[:,0],x_y_r[:,1],color='red', marker='o',
         linestyle='dashed',linewidth=0.1, markersize=1,label="y = 1/2*x + 1")
plt.legend()
plt.show()
```

```python
# 勾配のグラフ
plt.title("slope")
plt.xlabel("x")
plt.ylabel("slope")
plt.plot(x_y_r[:,0],koubai,color='red', marker='o',
         linestyle='dashed',linewidth=0.1, markersize=1,label="slope")
plt.legend()
plt.show()
```

### 【問題5】Pythonの関数化

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    変化量を使い勾配を計算する。

    Parameters
    ----------------
    function : function
    求めたい関数、xのndarrayを入れるとyのndarrayを返すもの。
    x_range : tuple
    np.arange()の引数と同じ方法で範囲を指定する。

    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
    xとyを結合したもの。nはx_rangeに依存する。
    gradient : ndarray, shape(n-1,)
    関数の勾配。差をとるため要素は1つ減る
    """

# 変数xにタプルで設定している範囲で配列を作成する(*〜)を使用すると
    array_x = np.arange(*x_range)
#     print(array_x)
# アトリビュートに対して配列の型を入れる(1001,1)
    array_x.shape = (*array_x.shape, 1)
#     print(array_x)
#     print(array_x.shape)
# 新しい配列xy に array_x と 出力される y を結合する
    array_xy = np.hstack((array_x, function(array_x)))
    print(array_xy)
# 変化量を計算
    delta = array_xy[1:] - array_xy[:-1]
#     print(delta)
# 勾配を計算
    gradient = delta[:, 1]/ delta[:, 0]
#     print(gradient)

    return array_xy, gradient

# 1/2*x + 1 の計算
# def function0(array_x):
#     array_y = 1/2*array_x + 1
#     return array_y

# x**2 の計算
def function1(array_x):
    array_y = array_x**2
    return array_y

# 2*x**2 + 2**x の計算
def function2(array_x):
    array_y = (2*array_x**2) + (2**array_x)
    return array_y

# # sin(x**(1/2))
def function3(array_x):
    array_y = np.sin(array_x**(1/2))
    return array_y

# array_xy, gradient = compute_gradient(function0)
array_xy, gradient = compute_gradient(function1)
# print(array_xy)
print(array_xy.shape)
print(array_xy.T)
print(array_xy.T.shape)
array_xy, gradient = compute_gradient(function2)
array_xy, gradient = compute_gradient(function3)
```

```python
array_xy, gradient = compute_gradient(function1)

plt.figure(dpi=90)
plt.title("y = x**2")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(array_xy.T[0], array_xy.T[1], label="y = x**2")
plt.legend()
plt.show()

plt.figure(dpi=90)
plt.title("y'")
plt.xlabel("x")
plt.ylabel("y'")
plt.plot(array_xy.T[0][1:],gradient, label="gradient")
plt.legend()
plt.show()
```

```python
array_xy, gradient = compute_gradient(function2)
print(array_xy.T[1][0:5])
plt.figure(dpi=90)
plt.title("y = 2*x**2 + 2^x")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(array_xy.T[0][0:600], array_xy.T[1][0:600], label="y = 2*x**2 + 2^x")
plt.legend()
plt.show()

plt.figure(dpi=90)
plt.title("y'")
plt.xlabel("x")
plt.ylabel("y'")
plt.plot(array_xy.T[0][1:601],gradient[0:600], label="gradient")
plt.legend()
plt.show()
```

```python
array_xy, gradient = compute_gradient(function2)
print(array_xy.T[1][0:5])
plt.figure(dpi=90)
plt.title("y = 2*x**2 + 2^x")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(array_xy.T[0], array_xy.T[1], label="y = 2*x**2 + 2^x")
plt.legend()
plt.show()

plt.figure(dpi=90)
plt.title("y'")
plt.xlabel("x")
plt.ylabel("y'")
plt.plot(array_xy.T[0][1:],gradient, label="gradient")
plt.legend()
plt.show()
```

```python
array_xy, gradient = compute_gradient(function3,x_range=(0, 50.1, 0.1))

plt.figure(dpi=90)
plt.title("y = sin(x**(1/2))")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(array_xy.T[0], array_xy.T[1], label="y = sin(x**(1/2))")
plt.legend()
plt.show()

plt.figure(dpi=90)
plt.title("y'")
plt.xlabel("x")
plt.ylabel("y'")
plt.plot(array_xy.T[0][1:],gradient, label="gradient")
plt.legend()
plt.show()
```
