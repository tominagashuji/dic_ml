【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
多次元の正規分布による乱数はnp.random.multivariate_normal()を使うことで作成できます。
作成したndarrayのshapeは(500, 2)になります。
シード値について
実行ごとに値が変わらないように、np.random.seed()により乱数のシード値を固定しておきましょう。今回シード値は0を使うことにします。
np.random.seed(0)

```python
"""
numpyを使うためにimportして npと言うインスタンスで使用する
提示されている平均で、２次元正規分布による乱数を500こ作成する
この時のndarrayの形は 500,2 にする
"""
import numpy as np

# 練習１
# npdata = np.random.seed(0)
# print(npdata)

# 練習2
# data = [[1.0, 2.0, 3.0], [11.0,12.0, 13.0]]
# npdate = np.array(data)
# print(npdate)
 
# 回答：
# 参考：http://starpentagon.net/analytics/numpy_norm_rand/

np.random.seed(0)

mu = [-3, 0] # 平均
sigma = [[1.0, 0.8], [0.8, 1.0]]#分散
 
# 2次元正規乱数を500個生成
values = np.random.multivariate_normal(mu, sigma, 500)
# print(values)
print(values.shape)
```

### 【問題2】散布図による可視化
問題1で作成したデータ点を散布図により可視化してください。
散布図はmatplotlibのplt.scatter()を使うことで描けます。
matplotlib.pyplot.scatter — Matplotlib 3.1.1 documentation

```python
"""
問題１の 二次元正規分布 を使用する
散布図を作成するには matplotlib を import する必要がある。
使い方はnumpyと同じ
https://pythondatascience.plavox.info/matplotlib/%E6%95%A3%E5%B8%83%E5%9B%B3
scatterと言うメソッドが散布図の様子
オプションは特にここでは取り扱わないよう進める
scatter で表示をする時に横と縦の値を用意するのだけれど、これの指定が分からない。
ここで使用している二次元正規分布の情報は２次元配列で、x軸には０列の情報を使いたく、y軸には１列の情報を使いたい。
https://qiita.com/supersaiakujin/items/d63c73bb7b5aac43898a
「:」が行列を固定させられるコマンドの様子。
"""
import matplotlib.pyplot as plt
# value_x = values
plt.scatter(values[:,0],values[:,1])
```

### 【問題3】ヒストグラムによる可視化

ヒストグラムはplt.hist()を使うことで描けます。

各次元ごとに作成し、2つのヒストグラムとしてください。
それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えるようにしてください。
matplotlib.pyplot.xlim — Matplotlib 3.1.1 documentation

```python
"""
ヒストグラムを作成する。方法はplt.hist()で可能。
各次元とあるが、、これは[:,0]と[:,1]で進める。
横軸の範囲も揃えるためには plt.xlim() を使用する必要がある。
"""
plt.hist(values[:,0])
plt.xlim(-6, 4)
plt.show()
plt.hist(values[:,1])
plt.xlim(-6, 4)
```

### 【問題4】データの追加
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題2のものは1を表示してください。

```python
"""
単純に問題１で作成した情報を values1 と言う名前で提示されている平均等を使って作成した。
それをscatterで同じグラフに書くことは出来た。特にオプションがなくても色分けがされて嬉しい。
ただ、表の右上の色事の名前？が出ないのでオプションを調査する必要がある。
https://own-search-and-study.xyz/2016/08/08/matplotlib-pyplot%E3%81%AEplot%E3%81%AE%E5%85%A8%E5%BC%95%E6%95%B0%E3%82%92%E4%BD%BF%E3%81%84%E3%81%93%E3%81%AA%E3%81%99/
label と言うオプションを使用する、また表示にはlegend()が必要とのこと。
"""

mu1 = [0, -3] # 平均
values1 = np.random.multivariate_normal(mu1, sigma, 500)
print(values1.shape)
# print(values1)

plt.scatter(values[:,0],values[:,1],label='0')
plt.scatter(values1[:,0],values1[:,1],label='1')
plt.legend()
```

### 【問題5】データの結合
データはまとめておいた方が後々扱いやすいです。
問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。
結合はnp.concatenate()やnp.vstack()を使うことで行えます。
numpy.concatenate — NumPy v1.16 Manual
numpy.vsplit — NumPy v1.16 Manual

```python
"""
https://docs.scipy.org/doc/numpy/reference/generated/numpy.concatenate.html
公式の example を使用したところ結合は簡単に出来た。
これは同型の配列をお尻にくっ付けている。
"""
values2 = np.concatenate((values,values1), axis = 0)
print(values2.shape)
print(values2)
```
