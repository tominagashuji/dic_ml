```python
"""
【課題1】1次元配列を作ろう
https://qiita.com/Y_F_Acoustics/items/4b1ee5c046870796b919
"""

FIRST = 0
ADD = 0.01
addition = ADD
array_list = [FIRST]

for i in range(100):
#     小数点を丸める方法 %g でどうにかする方法
#     print('%g' % addition)
#     array_list.append(float('%g' % addition)) #'%g'の余を求めると丸められる
#     addition += ADD

#     小数点を丸める方法 round で四捨五入する方法
    array_list.append(round(addition,2)) #'%g'の余を求めると丸められる
    addition += ADD

#     小数点を丸める方法 float を使用する方法もあるらしい。。

print(array_list)
```
