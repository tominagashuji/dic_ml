### 【問題2】ある地点の勾配を計算する関数を作成
「二次元配列と勾配問題」では各地点での勾配を最初に求めました。
しかし、暗闇の富士山を想像すると、知ることができるのは今いる地点の勾配だけです。
「今いる地点の番号」と「今いる隣の地点の番号（今いる地点の番号-1）」の関係による勾配を求める関数を作成してください。
勾配は以下の式で求まりましたが、このデータでは「標高」がy、「その地点の番号」がxに相当します。

```python
def get_slope(point):
    d = fuji[point] - fuji[point - 1]
    slope = d.T[3] / d.T[0]
    return slope
```

```python
point = 135
slope = get_slope(point)
print(slope)
```

### 【問題3】移動先の地点を計算する関数を作成
今いる地点の勾配の情報を元に、次に移動先の地点を計算する関数を作成してください。

```python
def next_point(i, alpha=0.2):
#     alpha = 0.2
#     print(fuji[i,0])
#     print(fuji[i,0] - alpha)
    # 今いる地点の勾配情報を元に移動先の地点を求める 
#     print(fuji.shape)
#     kakunin = alpha*get_slope(i)
#     print(kakunin.shape)
#     print(fuji[i,0].shape)
    n_point = fuji[i,0] - (alpha*get_slope(i)) # 次の地点は整数である必要がある、多分これだと次の地点は少数になる
#     print(n_point.shape)
    if n_point < 0:
#         print(n_point)
        print("その地点はマイナスです")
    elif n_point > len(fuji):
#         print(n_point)
        print("その地点はありません")
    else:
        # 四捨五入はdecimal & quantize を使用する事！
        n_point = Decimal(n_point).quantize(Decimal('0'), rounding=ROUND_HALF_UP)
#         print(n_point)
        return int(n_point) # decimal 型でreturnすると他でエラー発生
```

```python
# 今いる地点を適当に決める
point = 120
alpha = 0.2
# 次の地点を求める関数に今いる地点を投入すると、次の地点が求まる
n_point = next_point(point, alpha)
print(n_point)
```

```python
# 四捨五入学習
print(n_point) # 小数点含むデータを確認
print(Decimal(n_point)) # decimal データへ変換
print(type(Decimal(n_point))) # type 確認
# 四捨五入処理
print(Decimal(n_point).quantize(Decimal('0'), rounding=ROUND_HALF_UP))
```

### 【問題4】山を下る関数を作成
山を下る関数を作成してください。
最初の地点を受け取り、問題3の関数によって地点を更新していきます。
移動毎の地点はリストなどに記録して、これを関数の返り値としてください。

```python
"""
最初の地点を受け取り（point）、問題３の関数（next_point())を使って次の地点を求める。
次の地点に変化がなくなるまで繰り返しの実施をする。
尚、次の地点はリストで保存して関数を実行すると出力されるようにしよう。
"""
def down_fuji01(point=136):
    point_list = []
    point_list = [point]
#     print(point)
    n_point = next_point(point,alpha)
#     print(n_point)
    for i in range (10):
#         print(i)
#         print(point)
#         print(type(point))
#         print(point_list[i])
        point_list.append(next_point(point, alpha))
#         print(point_list[i+1])
        point = point_list[i+1]
#         print(point)
#         print(type(point))
#         if point_list[i] == point_list.append(next_point(point)):
        if point_list[i] == point_list[i+1]:
#             print(point)
#             print(point_list[i])
            return point_list
    return point_list
```

```python
point = 136

point_list = down_fuji01()
print(point_list)
```

```python
"""
上記のヤツだと最後が二重になって出力されるのでやめたい。
最初の地点を受け取り（point）、問題３の関数（next_point())を使って次の地点を求める。
次の地点に変化がなくなるまで繰り返しの実施をする。
尚、次の地点はリストで保存して関数を実行すると出力されるようにしよう。
"""
def down_fuji02(point=136):
    point_list = []
    point_list = [point]
    i = 0
    while True:
        n_point = next_point(point)
        if point_list[i] != n_point:
            point_list.append(n_point)
            point = point_list[i+1]
            i = i + 1
        else:
            break
    return point_list
```

```python
# point = 142
point_list = down_fuji02(point)
print(point_list)
```
